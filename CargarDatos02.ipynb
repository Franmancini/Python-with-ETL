```python
import pandas as pd

RUTA=r"C:\Users\NaruTovr_Lap\Downloads\clase7\fastfood.csv"
ARCHIVO=pd.read_csv(RUTA,sep=";")
ARCHIVO["sexo"]=ARCHIVO["sexo"].str[:1].str.upper()
ARCHIVO["satisfaccion"]=ARCHIVO["satisfaccion"].str[:1].astype(int)
ARCHIVO = ARCHIVO[["id", "sexo", "edad","satisfaccion"]]
FILTRO= ARCHIVO[ARCHIVO["satisfaccion"]==5]
print(FILTRO)
FILTRO.to_csv(r"C:\Users\NaruTovr_Lap\Downloads\clase8\ARCHIVO FINAL.csv",index=False)
```

# Envío de datos del dataframe a SQL Server
