# Conexión de datos desde un servidor SQL Server con la base de datos Northwind 

```python
import pyodbc
cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=hp;DATABASE=northwind;UID=sa;PWD=123456')
cursor = cnxn.cursor()
cursor.execute("select RegionID,RegionDescription from region")
rows = cursor.fetchall()
for row in rows:
    print(row.RegionID, row.RegionDescription)
```
